```python
# Feature Extraction Function
# Input shape is (n,x,y,x) - number of images, x, y, and channels

def feature_extractor(dataset):
    x_train = dataset
    image_dataset = pd.DataFrame()
    for image in range(x_train.shape[0]): 
        
        # Temp dataframe to capture info for each loop. Reset it to blank after each loop.
        df = pd.DataFrame()
        
        input_img = x_train[image, :, :, :]
        img = input_img
        print(input_img.shape)
        # Add feature extarction data to dataframe
        
        # Feature 1 - Pixel Value
        pixel_values = img.reshape(-1)
        df['Pixel_value'] = pixel_values # Pixel value itself as a feature
        print('pixel' + str(len(df['Pixel_value'])))
        
        # Feature 2 - Gabor filters
        
        num = 1 # To label different gabor filters
        kernels = []
        for theta in range(2): # define number of thetas
            theta = theta/4. * np.pi
            for sigma in (1, 3):
                for lamda in np.arange(0, np.pi, np.pi / 4): #Range of wavelengths
                    for gamma in (0.05, 0.5):
                        
                        gabor_label = 'Gabor' + str(num)
                        ksize = 9
                        kernel = cv2.getGaborKernel((ksize, ksize), sigma, theta, lamda, gamma, 0, ktype = cv2.CV_32F)
                        kernels.append(kernel)
                        
                        # Now filter the image and add values to a new col
                        fimg = cv2.filter2D(img, cv2.CV_8UC3, kernel)
                        print(len(fimg))
                        filtered_img = fimg.reshape(-1)
                        print(len(filtered_img))
                        df[gabor_label] = filtered_img # Labels columns as Gabor1, Gabor2, etc.
                        print(gabor_label)
                        num +=1 # Increment for gabor column label
        #print(len(df))
        
        # Feature 3 - Canny Edge
        '''
        thresh_sigma = 0.33
        #splitting image into respective 3 channels
        b,g,r = cv2.split(img)
        for i in (b,g,r):
            
            median = np.median(i)
            #print(median)
            lower = int(max(0, (1.0 - thresh_sigma) * median))
            upper = int(min(255, (1.0 + thresh_sigma) * median))
            i = cv2.Canny(i, lower, upper)
            #print(edges.shape)
        edges = cv2.merge((b,g,r))
        edges1 = edges.reshape(-1)
        #print(edges1.shape)
        df['Canny Edge'] = edges1
        '''
        thresh_sigma = 0.33
        #splitting image into respective 3 channels
        b,g,r = cv2.split(img)
        
        # Calculating median, lower and upper threshold for each channel
        median_b = np.median(b)
        lower_b = int(max(0, (1.0 - thresh_sigma) * median_b))
        upper_b = int(min(255, (1.0 + thresh_sigma) * median_b))

        median_g = np.median(g)
        lower_g = int(max(0, (1.0 - thresh_sigma) * median_g))
        upper_g = int(min(255, (1.0 + thresh_sigma) * median_g))

        median_r = np.median(r)
        lower_r = int(max(0, (1.0 - thresh_sigma) * median_r))
        upper_r = int(min(255, (1.0 + thresh_sigma) * median_r))
        
        # Applying Canny filter on each channel
        b = cv2.Canny(b, lower_b, upper_b)
        g = cv2.Canny(g, lower_g, upper_g)
        r = cv2.Canny(r, lower_r, upper_r)

        # Merging back the three channels
        edges = cv2.merge((b,g,r))
        edges1 = edges.reshape(-1)
        df['Canny Edge'] = edges1
        
        # Feature 4 - Roberts Edge
        b,g,r = cv2.split(img)

        b = roberts(b)
        g = roberts(g)
        r = roberts(r)
        
        edge_roberts = cv2.merge((b,g,r))
        edge_roberts1 = edge_roberts.reshape(-1)
        df['Roberts'] = edge_roberts1
        
        # Feature 5 - Sobel
        edge_sobel = sobel(img)
        edge_sobel1 = edge_sobel.reshape(-1)
        df['Sobel'] = edge_sobel1
        
        # Feature 6 - Scharr
        edge_scharr = scharr(img)
        edge_scharr1 = edge_scharr.reshape(-1)
        df['Scharr'] = edge_scharr1
        
        # Feature 7 - Prewitt
        edge_prewitt = prewitt(img)
        edge_prewitt1 = edge_prewitt.reshape(-1)
        df['Prewitt'] = edge_prewitt1
        
        # Feature 8 - Gaussian with sigma = 3
        from scipy import ndimage as nd
        gaussian_img = nd.gaussian_filter(img, sigma = 3)
        gaussian_img1 = gaussian_img.reshape(-1)
        df['Gaussian s3'] = gaussian_img1
        
        # Feature 9 - Gaussian with sigma = 5
        gaussian_img2 = nd.median_filter(img, size = 5)
        gaussian_img3 = gaussian_img2.reshape(-1)
        df['Gaussian s5'] = gaussian_img3
        
        # Feature 10 - Median with sigma = 3
        median_img = nd.median_filter(img, size = 3)
        median_img1 = median_img.reshape(-1)
        df['Median s3'] = median_img1
        
        image_dataset = image_dataset.append(df)
    
    return image_dataset
```
